```
import numpy as np
n=-1
x=2
while x > 1:
  n += 1
  x=(1+np.float16(1/(2**n)))
print('16 bits : n=',n,',x=',x) 
n=-1
x=2 
while x > 1:
  n += 1
  x=(1+np.float32(1/(2**n)))
print('32 bits : n=',n,',x=',x) 
n=-1
x=2
while x > 1:
  n += 1
  x=(1+np.float64(1/(2**n)))
print('64 bits : n=',n,',x=',x) 
n=-1
x=2
while x > 1:
  n += 1
  x=(1+np.float128(1/(2**n)))
print('128 bits :n=',n,',x=',x) 
```

```
import numpy as np
import matplotlib.pyplot as plot
time = np.arange(-3,3,0.1)
amplitude_sin = np.sin(time)
plot.plot(time, amplitude_sin)
plot.title('Sin and Tylor approx')
plot.xlabel('Time')
plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')
#plot.axhline(y=0, color='k')



def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)


tab_sin = []
for x in time:
  sinus = 0
  for n in range(2) :
    sinus = sinus + (((-1)**n)*((x)**(2*n+1))/(factorial(2*n+1)))
  tab_sin.append(sinus)

plot.plot(time, tab_sin, "r")
plot.show()
```

```
import numpy as np
import matplotlib.pyplot as plot
time = np.arange(-3,3,0.1)
amplitude_cos = np.cos(time)
plot.plot(time, amplitude_cos)
plot.title('Cos and Tylor approx')
plot.xlabel('Time')
plot.ylabel('Amplitude = cos(time)')
plot.grid(True, which='both')
#plot.axhline(y=0, color='k')



def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)


tab_cos = []
for x in time:
  cosinus = 0
  for n in range(2) :
    cosinus = cosinus + (((-1)**n)*((x)**(2*n))/(factorial(2*n)))
  tab_cos.append(cosinus)


plot.plot(time, tab_cos, "r")
plot.show()
```

```
difference_sin = tab_sin - amplitude_sin
plot.plot(time, difference_sin)
plot.title('Difference of sin and Tylor approx.')
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
#plot.axhline(y=0, color='k')
plot.axis([-3,3,-0.01,0.01])
plot.show()
```

```
difference_cos = tab_cos - amplitude_cos
plot.plot(time, difference_cos)
plot.title('Difference of cos and Tylor approx.')
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
#plot.axhline(y=0, color='k')
plot.axis([-3,3,-0.01,0.01])
plot.show()
```

```
import numpy as np
import matplotlib.pyplot as plot
time = np.arange(-3,3,0.1)
amplitude = np.sin(time)*np.sin(time)+np.cos(time)*np.cos(time)
plot.plot(time, amplitude)
plot.title('Sin^2+cos^2 and Tylor approx')
plot.xlabel('Time')
plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')



def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)


tab = []
for x in time:
  sinus = 0
  cosinus = 0
  for n in range(2) :
    sinus = sinus + (((-1)**n)*((x)**(2*n+1))/(factorial(2*n+1)))
    cosinus = cosinus + (((-1)**n)*((x)**(2*n))/(factorial(2*n)))
  tab.append(sinus*sinus+cosinus*cosinus)

plot.plot(time, tab, "r")
plot.axis([-3,3,0.6,1.4])
plot.show()
```

```
difference = tab - amplitude
plot.plot(time, difference)
plot.title('Difference  and Tylor approx.')
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
#plot.axhline(y=0, color='k')
plot.axis([-3,3,-0.04,0.04])
plot.show()
```
